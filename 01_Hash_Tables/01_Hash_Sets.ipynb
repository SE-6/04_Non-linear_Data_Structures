```python
class HashSet:
    def __init__(self, size=10) -> None:
        self.size = size
        self.buckets = [[] for _ in range(size)]

    def _hash(self, key):
        return hash(key) % self.size

    def add(self, key):  # O(1), but deteriorates to O(n)
        index = self._hash(key)  # O(1)
        bucket = self.buckets[index]
        if key not in bucket:  # O(1), but deteriorates to O(n)
            bucket.append(key)

    def remove(self, key):  # O(1), but deteriorates to O(n)
        index = self._hash(key)
        bucket = self.buckets[index]
        try:
            bucket.remove(key)  # O(n)
        except:
            pass

    def contains(self, key):  # O(1), but deteriorates to O(n)
        index = self._hash(key)
        bucket = self.buckets[index]
        return key in bucket  # O(n)
```

```python
my_hash_set = HashSet()

print(my_hash_set.buckets)
my_hash_set.add("Ali")
my_hash_set.add("Leila")
my_hash_set.add("Andrew")
print(my_hash_set.buckets)
my_hash_set.remove("Not in there")
print(my_hash_set.buckets)
print(my_hash_set.contains("Ali"))
print(my_hash_set.contains("Jorge"))
```
