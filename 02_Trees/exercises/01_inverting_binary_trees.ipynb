![invert bin tree](https://assets.leetcode.com/uploads/2021/03/14/invert1-tree.jpg)


```python
# 1. Approach: Iterative with Breadth First Traversal


def invertTree(root):
    if root is None:
        return root

    # BFS: traverse the tree node for node
    # process node and put children on stack
    stack = [root]

    while stack:
        curr = stack.pop()  # get top item (LIFO)
        curr.left, curr.right = curr.right, curr.left  # swap left-right
        # Put possible children on stack
        if curr.right is not None:
            stack.append(curr.right)
        if curr.left is not None:
            stack.append(curr.left)
    return root
```

```python
# 2. Approach: Recursion
# (very compact version)
def invertTree(self, root):
    if root:
        # perform invertTree on each child and reassign in swapped order
        root.left, root.right = self.invertTree(root.right), self.invertTree(root.left)
    return root
```
