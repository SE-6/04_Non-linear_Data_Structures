```python
map_ = """..C......
.........
....m...."""
moves = 5


def cat_mouse(map_, moves):
    threeD = map_.split("\n")
    print(threeD)
    cat_position = None
    mouse_position = None

    for row_index, row in enumerate(threeD):
        for char_index, character in enumerate(row):
            if character == "C":
                cat_position = [row_index, char_index]
            if character == "m":
                mouse_position = [row_index, char_index]

    if cat_position is None or mouse_position is None:
        return "boring without two animals"

    cat_row, cat_line = cat_position
    mouse_row, mouse_line = mouse_position

    distance = abs(cat_row - mouse_row) + abs(cat_line - mouse_line)

    if distance <= moves:
        return "Caught!"
    else:
        return "Escaped!"


cat_mouse(map_, moves)
```

```python
def cat_mouse(map_, moves):
    if "C" not in map_ or "m" not in map_:
        return "boring without two animals"
    rows = map_.split("\n")
    c_pos = None
    m_pos = None
    for i, row in enumerate(rows):
        if "C" in row:
            c_pos = (i, row.index("C"))
        if "m" in row:
            m_pos = (i, row.index("m"))
    dis = abs(c_pos[0] - m_pos[0]) + abs(c_pos[1] - m_pos[1])
    if moves >= dis:
        return "Caught!"
    return "Escaped!"
```
