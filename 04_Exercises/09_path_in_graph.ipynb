```python
class Graph:
    def __init__(self, edges):
        self.edges = edges
        self.create_adj_dict()

    def create_adj_dict(self):
        self.adj_dict = {}
        for start, dist in self.edges:
            if start in self.adj_dict:
                self.adj_dict[start].append(dist)
            else:
                self.adj_dict[start] = [dist]
            if dist in self.adj_dict:
                self.adj_dict[dist].append(start)
            else:
                self.adj_dict[dist] = [start]

    def has_path(self, start, destination):
        stack = [start]
        visited = {start}
        while stack:
            current_node = stack.pop()
            if current_node == destination:
                return True
            for neighbor in self.adj_dict.get(current_node, []):
                if neighbor not in visited:
                    visited.add(neighbor)
                    stack.append(neighbor)
        return False
```

```python
def is_valid_path(edges: list[list[int]], source=0, destination=2) -> bool:
    if not edges:
        return False

    if source == destination:
        return True

    # queue = list(filter(lambda edge: any([edge[0] == source, edge[1] == source]), edges))
    queue = [edge for edge in edges if edge[0] == source or edge[1] == source]

    is_valid = False

    visited_vertices = set([source])

    while queue:
        current_edge = queue.pop(0)

        if current_edge[0] == destination or current_edge[1] == destination:
            is_valid = True
            break

        neighbor = None
        if current_edge[0] not in visited_vertices:
            neighbor = current_edge[0]
        elif current_edge[1] not in visited_vertices:
            neighbor = current_edge[1]

        if neighbor is not None:
            visited_vertices.add(neighbor)
            queue.extend(
                [
                    edge
                    for edge in edges
                    if any([edge[0] == neighbor, edge[1] == neighbor])
                    and (
                        edge[0] not in visited_vertices
                        or edge[1] not in visited_vertices
                    )
                ]
            )

    return is_valid


test1 = is_valid_path([[0, 1], [1, 2], [2, 3]], 0, 3)  # Output: True
test2 = is_valid_path(edges=[[0, 1], [1, 2], [2, 0]])  # Output: True
test3 = is_valid_path(
    edges=[[0, 1], [0, 2], [3, 5], [5, 4], [4, 3]], source=0, destination=5
)  # Output: False

print(test1)
print(test2)
print(test3)
```

```python
from collections import deque

n = 3
edges = [[0, 1], [1, 2], [2, 0]]
source = 0
destination = 2
# expected = True


def validPath(
    n: int, edges: list[list[int]], source: int, destination: int
) -> bool:  #  O(n + edges), because of visited list
    graph = {}

    for i in range(n):
        graph[i] = []

    for node_a, node_b in edges:
        graph[node_a].append(node_b)
        graph[node_b].append(node_a)

    visited = [False] * n
    visited[source] = True

    queue = deque([source])

    while queue:  # O(vertices)
        curr_node = queue.popleft()
        if curr_node == destination:
            return True

        neighbors = graph[curr_node]

        for node in neighbors:  # O(edges)
            if not visited[node]:
                queue.append(node)
                visited[node] = True

    return False


validPath(n, edges, source, destination)
```
