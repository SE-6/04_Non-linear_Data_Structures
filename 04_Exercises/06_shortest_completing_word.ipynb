```python
"""
You are given a string, licensePlate, and an array of strings, words. Your task is to find the shortest word in words that completes the licensePlate.

A word is considered completing if it contains all the letters from licensePlate, ignoring digits and spaces, and treating uppercase and lowercase letters as the same. If a letter appears multiple times in licensePlate, it must appear at least that many times in the word.

You can assume that at least one completing word will always exist. If there are multiple words of the same shortest length, return the one that appears first in the list.

https://leetcode.com/problems/shortest-completing-word/description/
"""
```

```python
licensePlate = "1s3 PSt"
words = ["step", "steps", "stripe", "stepple"]
expected = "steps"

# licensePlate = "Ah71752"
# words = [
#     "suggest",
#     "letter",
#     "of",
#     "husband",
#     "easy",
#     "education",
#     "drug",
#     "prevent",
#     "writer",
#     "old",
# ]


def shortestCompletingWord(
    licensePlate: str, words: list[str]
) -> str | None:  # O(n² + m * k)
    result = None
    licensePlate = licensePlate.lower()

    license = {
        char: licensePlate.count(char) for char in set(licensePlate) if char.isalpha()
    }  # O(licensePlate²)

    for word in words:  # O(words)

        lower_word = word.lower()
        if all(
            lower_word.count(char) >= license.get(char, 0)
            for char in license.keys()  # O(word * licensePlate)
        ):
            if result and len(word) > len(result):
                continue
            result = word

    return result


shortestCompletingWord(licensePlate, words)
```

```python
def shortestCompletingWord(licensePlate: str, words: list[str]) -> str:  # O(ws * (l²))
    filtered_words = []

    for word in words:  # O(words)
        word_lower = word.lower()
        license_plate_lower = licensePlate.lower()
        is_completing = True

        for char in license_plate_lower:  # O(licensePlate)
            if char.isalpha():
                count_in_word = word_lower.count(char)  # O(word)
                count_in_plate = license_plate_lower.count(char)  # O(licensePlate)
                if count_in_word < count_in_plate:
                    is_completing = False
                    break

        if is_completing:
            filtered_words.append(word)

    filtered_words.sort(key=len)  # O(n log n)
    return filtered_words[0]


shortestCompletingWord(licensePlate, words)
```

```python
def get_freq(s: str) -> list[int]:
    freq = [0] * 26
    for char in s.lower():
        if "a" <= char <= "z":
            index = ord(char) - ord("a")
            freq[index] += 1
    return freq


def shortestCompletingWord(
    licensePlate: str, words: list[str]
) -> str | None:  # O(licensePlate + words * word length)

    target_freq = get_freq(licensePlate)
    shortest = None

    for word in words:
        if shortest and len(word) >= len(shortest):
            continue

        word_freq = get_freq(word)

        is_match = True
        for i in range(len(target_freq)):
            if word_freq[i] < target_freq[i]:
                is_match = False
                break

        if is_match:
            shortest = word

    return shortest


shortestCompletingWord(licensePlate, words)
```
