```python
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right
```

```python
preorder = [8, 5, 1, 7, 10, 12]


def bstFromPreorder(preorder: list[int]):
    # A pointer to track which number in the list is currently processed
    index = 0

    # Helper function that sets 'boundaries' for what numbers are allowed in a specific branch
    def build(lower=float("-inf"), upper=float("int")):
        # Allows this inner function to update the 'index' variable defined in the outer function
        nonlocal index

        # Base Case: we've processed every number in the list, stop and return nothing
        if index == len(preorder):
            return None

        # Get the current number from the list to try and make a node
        value = preorder[index]

        # BST Rule Check: If the number doesn't fit within the current min/max bounds, it belongs elsewhere
        if value < lower or value > upper:
            return None

        # Found a valid spot for this number, so move the pointer forwards
        index += 1
        # Create the node
        node = TreeNode(value)

        # Build the left side: Any numbers here are smaller than the current node's value
        node.left = build(lower, value)
        # Build right side: Any numbers here are larger than current
        node.right = build(value, upper)

        # Return the finished node (and its attached sub-trees) to the caller
        return node

    # Kick off the process with no boundaries (infinity), starting at the root
    return build()


print(bstFromPreorder(preorder))
```

```python
# Iterative approach using a stack to keep track of parent nodes
def bstFromPreorder(preorder: list[int]) -> TreeNode | None:

    # Initialize the root of the tree with first value in the list
    root = TreeNode(preorder[0])
    # The stack helps us keep track of nodes we might still need to attach children to
    stack = [root]
    # Start looking at the second element in the list (index 1)
    i = 1
    # This will store the last node popped from the stack (potential parent)
    last = None

    # Loop through the remaining elements in preorder list
    while i < len(preorder):
        # Create a new node for the current number
        new_node = TreeNode(preorder[i])

        # If the new value is smaller than the top of the stack, it's a left child
        if new_node.val < stack[-1].val:
            # Connect it to the left side of the current top node
            stack[-1].left = new_node
            # Add this new node to the stack to check for its own children later
            stack.append(new_node)
        else:
            # If the value is larger, we need to find its correct parent up the tree
            # Pop nodes off the stack until we find the one this new_node belongs to ('backtracking')
            last = stack.pop()
            # Replacing last with top of the stack until we found the right one
            while stack and new_node.val > stack[-1].val:
                last = stack.pop()

            # new_node becomes the right child of the last popped node
            last.right = new_node
            # Add the new node to the stack as it might have its own children
            stack.append(new_node)

        # Increment the index to move to the next number in the list
        i += 1

    # Return the completed root node
    return root
```
