```python
"""
You are given an array of strings, words. Your task is to return the list of words that can be typed using letters from only one row of an American keyboard.

Each letter is case-insensitive, meaning uppercase and lowercase versions of the same letter belong to the same row.

https://leetcode.com/problems/keyboard-row/description/
"""
```

```python
words = ["Hello", "Alaska", "Dad", "Peace"]
expected = ["Alaska", "Dad"]
```

```python
def findWords(words: list[str]) -> list[str]:  # O(n * 3k) -> O(n * k)
    output = []
    for word in words:  # O(n)
        lower_word = word.lower()
        if (
            all(char in "qwertyuiop" for char in lower_word)  # O(k)
            or all(char in "asdfghjkl" for char in lower_word)  # O(k)
            or all(char in "zxcvbnm" for char in lower_word)  # O(k)
        ):
            output.append(word)

    return output


findWords(words)
```

```python
set1 = set("qwertyuiop")
set2 = set("asdfghjkl")
set3 = set("zxcvbnm")


def findWords(words: list[str]) -> list[str]:  # -> O(n * k)

    result = []

    for word in words:  # O(n)
        lower_word = set(word.lower())
        for s in (set1, set2, set3):  # O(1)
            if lower_word.issubset(s):  # O(k)
                result.append(word)
                break

    return result


findWords(words)
```

```python
row1 = set("qwertyuiopQWERTYUIOP")
row2 = set("asdfghjklASDFGHJKL")
row3 = set("zxcvbnmZXCVBNM")


def findWords(words: list[str]) -> list[str]:
    out = []

    for word in words:
        word_set = set(word)
        if (
            word_set.issubset(row1)
            or word_set.issubset(row2)
            or word_set.issubset(row3)
        ):
            out.append(word)

    return out


findWords(words)
```

```python
def is_valid(word):
    word = word.lower()
    rows = ["qwertyuiop", "asdfghjkl", "zxcvbnm"]

    for row in rows:  # O(1)
        for letter in word:  # O(k)
            if letter not in row:
                break
        else:
            return True

    return False


def findWords(words: list[str]) -> list[str]:  # O(n * k)
    result = []
    for word in words:  # O(n)
        if is_valid(word):
            result.append(word)

    return result


findWords(words)
```
