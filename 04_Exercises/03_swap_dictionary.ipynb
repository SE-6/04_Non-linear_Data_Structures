```python
dictionary = {"Ice": "Cream", "Age": "21", "Light": "Cream", "Double": "Cream"}
expected = {"Cream": ["Ice", "Double", "Light"], "21": ["Age"]}


def switch_dict(dic):
    result = {}

    for key, value in dic.items():
        if value not in result:
            result[value] = []
        result[value].append(key)

    return result


def switch_dict(dic):
    result = {}

    for key, value in dic.items():
        entry = result.get(value, [])
        entry.append(key)
        result[value] = entry

    return result


def switch_dict(dic):
    result = {}

    for key, value in dic.items():
        result.setdefault(value, []).append(key)

    return result


switch_dict(dictionary)
```
